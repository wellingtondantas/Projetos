```python
#Libs
import pandas as pd

import seaborn as sns
```

```python
#ler o dataset
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", sep=",")
```

```python
#Muda o nome das colunas
df.columns = [ 'Class'
                 ,'alcohol'
                 ,'malicAcid'
                 ,'ash'
                 ,'ashalcalinity'
                 ,'magnesium'
                 ,'totalPhenols'
                 ,'flavanoids'
                 ,'nonFlavanoidPhenols'
                 ,'proanthocyanins'
                 ,'colorIntensity'
                 ,'hue'
                 ,'od280_od315'
                 ,'proline'
             ]
```

```python
#Mostra os dados
df
```

```python
#Estatística descritiva
df.describe()
```

```python
#Seleciona algumas colunas
df_sub = df[['proline','alcohol','malicAcid','magnesium','flavanoids']]
```

```python
#Plota boxplot de algumas colunas
ax = sns.boxplot(data=df_sub)
```

```python
#Estatística descritiva das colunas selecionadas
df_sub.describe()
```

```python
#Distribuição da coluna alcohol
ax = sns.distplot(df_sub.alcohol)
```
