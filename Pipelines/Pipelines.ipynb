# Bibliotecas

```python
#Libs Gerais
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
#from sklearn.preprocessing import OneHotEncoder

import category_encoders as ce

import pandas as pd

#Libs de modelos de ML
from sklearn import tree
```

# Lendo Dataset

```python
colunas = ['age', 'workclass','fnlwgt','education', 
           'educational-num', 'marital-status', 'occupation', 
           'relationship', 'race', 'gender', 'capital-gain', 
           'capital-loss', 'hours-per-week', 'native-country', 'income']
```

```python
df = pd.read_csv('resources/adult.data', names = colunas)
df
```

# 1° Parte: Preparação/ Manipulação de Dados e Pré-processamento

```python
#Verifica valores faltantes
df.isnull().sum()
```

```python
#Remove colunas desnecessárias para análise
df.drop(['education'], axis=1, inplace=True)
```

```python
#Separa os atribulos dos rotulos
X = df.drop(['income'], axis=1, inplace=False)
y = df['income']
```

```python
#Separa os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.2)
```

```python
#Verifica colunas numéricas e categoricas
df.select_dtypes(include='object')
```

```python
#Aplica o One Hot Encoder
ohe = ce.OneHotEncoder(use_cat_names=True)
```

```python
X_train = ohe.fit_transform(X_train)
```

```python
X_train
```

```python
#Aplica o pre-processador
scaler = StandardScaler().fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

## Avaliação de Modelo

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

#### Avaliação do conjunto de treinamento 

```python
#Retorna com os valores de predição do conjunto de treinamento
pred = model.predict(X_train)
```

```python
#Media de acerto do modelo para o conjunto de treinamento
media = accuracy_score(pred, y_train)
media
```

#### Avaliação do conjunto de teste 

```python
X_test = ohe.transform(X_test)
```

```python
scaler_test = StandardScaler().fit(X_test)
```

```python
X_test = scaler_test.transform(X_test)
```

```python
pred_test = model.predict(X_test)
```

```python
media_test = accuracy_score(pred_test,y_test)
media_test
```
