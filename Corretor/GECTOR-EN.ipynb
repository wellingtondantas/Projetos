```python
from gector.gec_model import GecBERTModel

model = GecBERTModel(vocab_path = "/content/gector/data/output_vocabulary", model_paths = ["/content/gector/roberta_1_gector.th"])
```

```python
def check_grammar(sent):
  predictions = []
  cnt_corrections = 0
  batch = []

  batch.append(sent.split())
  if len(batch) == 32:
    preds, cnt = model.handle_batch(batch)
    predictions.extend(preds)
    cnt_corrections += cnt
    batch = []
  if batch:
    preds, cnt = model.handle_batch(batch)
    predictions.extend(preds)
    cnt_corrections += cnt
  return " ".join(preds[0])
```

```python
sent = 'goot morning the boys is playing'
```
